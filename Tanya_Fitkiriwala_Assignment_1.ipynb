```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
excel_file = "C:/Users/tanya/Downloads/big_mac_jul_2015-1.xlsx"  
data = pd.read_excel(excel_file)
```

```python
data
```

```python
x = data['1000usd']
y = data['dollar_price']

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
```

```python
plt.scatter(x, y, label='Data Points')

regression_line = slope * x + intercept

reg_eq = f'Regression Equation: y = {slope:.4f}x + {intercept:.4f}'

plt.plot(x, regression_line, color='red', label='Regression Line')
plt.xlabel('GDP per Person, 2014, 1000s USD')
plt.ylabel('Big Mac Price, USD (At Market Exchange Rate)')
plt.title('July 2015 Big Mac Prices n=49 Countries')
plt.text(0.1, 0.1, reg_eq, fontsize=12, color='blue')
```

```python
print("Regression Values:")
print(f"Slope (m): {slope:.4f}")
print(f"Intercept (b): {intercept:.4f}")
print(f"Correlation Coefficient (r-value): {r_value:.4f}")
print(f"P-Value: {p_value:.4f}")
print(f"Standard Error: {std_err:.4f}")
```
