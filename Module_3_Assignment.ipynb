The dataframe `a_df` loads the dataset that contains information on a number of categories regarding traffic (variable `cnt` represents the actual traffic) at US airports from February 2011.
The dataset has already been loaded for you in the first code cell above.
 1. Print the number of rows for the subset of the dataframe for which traffic values are higher than `10,000` (ideally, you should format the print to put the thousands comma separator--as in `23,400` instead of `23400`
 2. Print the names of all the columns of the dataframe
 3. Print a subset of the dataframe containing the `10th` and the `3rd` rows of the original dataframe (`in that order`).
 4. Which US `state` has the `8th` largest __total__ airport traffic? Please, make sure you use the correct variable as described above.
 5. Which state has the second-most __total__ number of airports (i.e. the second-most amount of airports)?
 6. Of the cities that have multiple airports, which one has the least total traffic for all airports?
 7. What is the total airport traffic for all states that border the state of Massachusetts?


```python
# YOUR SECTION B ANSWER HERE (please use comments to number each of your answers)

# 1. Print the number of rows for the subset of the dataframe for which traffic values are higher than `10,000`
print(f'[1] Row with traffic vals > 10,000: {len(a_df.loc[a_df["cnt"] > 10_000]):,}\n')

# 2. Print the names of all the columns of the dataframe
print(f'[2] Column names of a_df dataframe: {a_df.columns.values}\n')

# 3. Print a subset of the dataframe containing the `10th` and the `3rd` rows of the original dataframe (`in that order`).
print(f'[3] 10th and 3rd row of a_df dataframe:\n{a_df.iloc[[10 - 1, 3 - 1]]}\n')

# 4. Which US `state` has the `8th` largest total airport traffic? Please, make sure you use the correct variable as described above.
print(f'[4] State with 8th highest total traffic: {(a_df.groupby("state")["cnt"].sum()).nlargest(8)[[-1]].index[0]}\n')

# 5. Which state has the second-most total number of airports (i.e. the second-most amount of airports)?
print(f'[5] State with 2nd highest total no. of airports: {a_df["state"].value_counts().nlargest(2)[[-1]].index[0]}\n')

# 6. Of the cities that have multiple airports, which one has the least total traffic for all airports?

# Creating a subset with aggregate functions to count the number of  size and sum
airports_traffic_by_city = a_df.groupby('city').agg(
    tot_airports=('city', 'size'),
    tot_traffic=('cnt', 'sum')
)
# To find the city with minimum traffic out of all the cities that have more than 1 airports in it.

airport_city_traffice_least = airports_traffic_by_city.loc[airports_traffic_by_city['tot_airports'] > 1].idxmin()['tot_traffic']
print(f'[6] City with total airport traffic which has least is : {airport_city_traffice_least}')

# 7. What is the total airport traffic for all states that border the state of Massachusetts?
# States bordering Massachusetts(Ref: https://thefactfile.org/u-s-states-and-their-border-states/)
states_bordering_MA_tot = a_df.loc[a_df['state'].isin(['NY', 'RI', 'VT','CT', 'NH',])]['cnt'].sum()
print(f'[7] Total traffic for all states bordering Massachusetts: {states_bordering_MA_tot:,}')
```
